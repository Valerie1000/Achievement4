# Check dtypes

```python
df_ords_prods.info()
```

```python
# Change types for df_ords_prods data set

df_ords_prods['order_id']=df_ords_prods['order_id'].astype('int32')
df_ords_prods['user_id'] = df_ords_prods['user_id'].astype('int32')
df_ords_prods['order_number']=df_ords_prods['order_number'].astype('int8')
df_ords_prods['orders_day_of_week']=df_ords_prods['orders_day_of_week'].astype('int8')
df_ords_prods['order_hour_of_day']=df_ords_prods['order_hour_of_day'].astype('int8')
df_ords_prods['days_since_prior_order']=df_ords_prods['days_since_prior_order'].astype('float16')
df_ords_prods['product_id']=df_ords_prods['product_id'].astype('int32')
df_ords_prods['add_to_cart_order']=df_ords_prods['add_to_cart_order'].astype('int32')
df_ords_prods['reordered']=df_ords_prods['reordered'].astype('int8')
```

```python
nan_percent = (df_ords_prods['product_id'].isna().mean()) * 100
print(f"Missing values in 'product_id': {nan_percent:.2f}% of total rows.")
```

```python
df_ords_prods.info()
```

```python
df_products_checked.info()
```

```python
# Change types for df_products_checked data set 

df_products_checked['product_id'] =df_products_checked['product_id'].astype('int32')
df_products_checked['aisle_id']=df_products_checked['aisle_id'].astype('int16')
df_products_checked['department_id']=df_products_checked['department_id'].astype('int32')
df_products_checked['prices']=df_products_checked['prices'].astype('float16')
```

```python
# Attemp a merge

df_ords_prods_merge = df_products_checked.merge(df_ords_prods, on = ['product_id'], how = 'inner', indicator = True)
```
