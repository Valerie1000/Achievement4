# Create a new column max_order that will be defined by grouping the user_id with the max count of order_number - the max number that each customer has ordered

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.shape
```

# Assign loyalty values to the max_order column where if the max is above 40 the string is 'loyal customer', if more than ten and equal/less than 40 then 'regular customer', and if less than 10 then 'new customer"

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

# The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

## Create new column average_spent

```python
df_ords_prods_merge['average_spent'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
df_ords_prods_merge.head()
```

## Assign low spender and high spender values

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_spent']<10,'spending_flag']='Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_spent']>= 10, 'spending_flag'] = 'High spender'
```

## Get counts

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

## Create column median_days_since_orders

```python
df_ords_prods_merge['median_days_since_orders'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Check head
df_ords_prods_merge.head()
```

##  Assign order_frequency_flag

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_orders'] > 10) & (df_ords_prods_merge['median_days_since_orders'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

## Check counts

```python
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
