# Define price_label to then flag a column labeled price_range

```python
# Define the function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# 'Apply the function
df['price_range']=df.apply(price_label, axis=1)
```

```python
# Check counts for each return in price_range
df['price_range'].value_counts(dropna=False)
```

```python
#Check max price in subset since there are supposedly zero counts for high range but my results do show them
df['prices'].max()
```

## Create a list 'results' to loop through orders-of-the-week and return strings.  Busiest_day is 0. Next_busiest_day is 1.  Slowest_day is 4 and Next_slowest_day is 3.  

```python
results = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    results.append("Busiest day")
  elif value == 4:
    results.append("Least busy")
  elif value == 1:
     results.append("Next_busiest_day")
  elif value == 3:
     results.append("Next_slowest_day")
  else:
    results.append("Regularly busy")
```

```python
df_ords_prods_merge['busiest_days']=results
```

# Create a list 'busy_hour' to loop through orders-of-the-week and return strings. 
### Most_orders will include hours 10 down to 12  - from list above
### Average_orders will include from hours 16 down to 19
### Fewest_orders will include from hourse 20 down to 3

```python
busy_hour = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12]:
    busy_hour.append("Most orders")
  elif value in [19,18,8,17,9,16]:
    busy_hour.append("Average orders")
  elif value in[3,4,2,5,1,0,6,23,22,21,7,20]:
    busy_hour.append("Fewest orders")

  else:
    busy_hour.append("No information")
      
```

## Use the list on df and get frequency on the new column

```python
df_ords_prods_merge['busiest_period_of_day'] = busy_hour
```
